```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.style.use('ggplot')

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

explode = [0,0,0,0.1,0]

plt.pie(slices,
        labels=labels,
        wedgeprops={'edgecolor':'black','linewidth':2}, # to alter the aesthetic properties of the chart
        explode = explode, # explodes the slices according to the explode list's values
        shadow = True, # adds shadows to the slices
        startangle=90, # specifies the angle the chart should start with
        autopct = '%1.1f%%' # it adds percentage value to the slices
       )
plt.show()
```
