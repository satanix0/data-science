```python
fig = plt.figure()

# Large Axes
laxes = fig.add_axes([0,0,1,1])
laxes.plot(a,b,label='ab')

laxes.set_xlim(0,8)
laxes.set_ylim(0,8000)
laxes.set_title('Large Plot')
laxes.set_xlabel('X values')
laxes.set_ylabel('Y values')

# Small Axes
saxes = fig.add_axes([0.2,0.5,0.3,0.3])
saxes.plot(a,b,label='ab')

saxes.set_xlim(1,2)
saxes.set_ylim(0,50)
saxes.set_title('Small Plot')
saxes.set_xlabel('A values')
saxes.set_ylabel('B values')
saxes.legend()

plt.show()
```

```python
fig=plt.figure(figsize=(5,4),dpi=100,facecolor='green',edgecolor='red')
ax1 = fig.add_axes([0,0,1,1])
ax1.plot(a,b)
ax1.set_ylim(0,8000)
ax1.set_ylabel('Y vales')

fig.savefig('newfig.png',bbox_inches='tight')
```
