```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,11,10)
x
```

```python
fig = plt.figure(figsize=(4,3))
ax1= fig.add_axes([0,0,1,1])
ax1.plot(x,x,label='x vs x',color='black',lw=1,ls='--',marker='v')
ax1.plot(x,x**2,label='x vs x squared',marker='o')
ax1.plot(x,x+10,label='x vs x+10')
plt.xkcd()
ax1.legend()

plt.savefig('innerplt.png',bbox_inches='tight')
```
