```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a= np.linspace(0,10,11)
```

```python
b = a**4
```

```python
x = np.arange(0,10)
y = 2*x
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[0][0].set_title('XY Plot')

axes[0][1].plot(a,b)
axes[1][0].plot(x,y**y)
axes[1][1].plot(a,b**2)

fig.suptitle(t="Figure Level Suptitle",fontsize=15)

fig.subplots_adjust(wspace = 1, hspace = 0.5)

fig.savefig('new_subplots.png',bbox_inches='tight')
```
