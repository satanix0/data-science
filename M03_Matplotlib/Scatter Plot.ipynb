```python
x=[5,7,8,5,6,7,9,2,3,4,4,4,2,6,3,6,8,6,4,1]
y=[7,4,3,9,1,3,2,5,2,4,8,7,1,6,4,9,7,7,5,1]
colors = [9,2,3,4,1,4,3,2,5,2,4,8,7,1,7,8,5,6,7,9,]
sizes = [random.randrange(100, 600, 1) for i in range(20)]

plt.scatter(x,y,
            s=sizes, # controls size of the points
            c=colors,edgecolor='black', # colors controls
            linewidth=1,alpha=1, # edge width and opacity
            cmap='OrRd' # color palette used  
           )

cbar = plt.colorbar() # create a indicating colorbar
cbar.set_label('Customer Satisfaction')
```

## importing data from github raw file.

```python
import requests
url='https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/master/Python/Matplotlib/07-ScatterPlots/2019-05-31-data.csv'

response = requests.get(url = url,
                        allow_redirects= True
                       )
with open('YT_data.csv','wb') as fp:
    fp.write(response.content)
    
data = pd.read_csv('YT_data.csv')
data
```

```python
plt.scatter(data['view_count'],data['likes'],
            c= data['ratio'],
            cmap='summer',
           linewidth=1,edgecolor='black',
           )
plt.xscale('log')
plt.yscale('log')

cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')
plt.tight_layout()
```
