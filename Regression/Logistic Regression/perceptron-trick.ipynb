```python
from sklearn.datasets import make_classification
import numpy as np

# Creatingb a classification DataSet
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=10)
```

```python
X.shape, y.shape
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```

```python
def step(z):
    return 1 if z > 0 else 0

def perceptron(X,y):
    # inserting 1 to all rows in the 0th Col - bias term (Xo)
    X = np.insert(X,0,1,axis=1)
    # initializing all the weights(coeffs) with 1s
    weights = np.ones(X.shape[1])
#     setting a learning rate
    lr = 0.1
    
    # 1000 epochs
    for i in range(1000):
        # picking a random datapoint/index 
        j = np.random.randint(0,100)
        # predicting for the chosen index
        y_hat = step(np.dot(X[j],weights))
        # updating weight by W(new) = W(old) + lr*(y[j]-y_hat[j])*X[j]
        
        weights = weights + lr*(y[j]-y_hat)*X[j]
        
#         returning bias(Xo),other coeffs
    return weights[0],weights[1:]
        
```

```python
intercept_,coef_ = perceptron(X,y)
```

```python
# Values with respect to Ax + By + C = 0
# [A,B]
print(coef_)
# C 
print(intercept_)
```

```python
# m = -(A/B), b = -C/B

m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```

```python
def perceptron(X,y):
    
    m = []
    b = []
    
    X = np.insert(X,0,1,axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.1
    
    for i in range(200):
        j = np.random.randint(0,100)
        y_hat = step(np.dot(X[j],weights))
        weights = weights + lr*(y[j]-y_hat)*X[j]
        
        m.append(-(weights[1]/weights[2]))
        b.append(-(weights[0]/weights[2]))
        
    return m,b
```

```python
m,b = perceptron(X,y)
```
