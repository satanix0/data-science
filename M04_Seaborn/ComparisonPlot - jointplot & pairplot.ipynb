# 1. jointplot(x,y,data,kind,alpha........)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')

df = pd.read_csv('..\\Datasets\\StudentsPerformance.csv')
```

```python
df.head()
```

```python
sns.jointplot(data =df,x='writing score',y='reading score', 
              hue='gender'
             )
```

```python
sns.jointplot(data =df,x='writing score',
              y='reading score',
              kind='scatter',
              alpha=0.2

)
```

####  Note: Available plot kinds are:
            1. scatter
            2. kde
            3. reg
            4. hist
            5. hex
            6. resid

```python
sns.jointplot(data =df,x='writing score',
              y='reading score',
              kind='hist',
              space=0.6 # oversees space between plots
)
```

```python
sns.jointplot(data =df,x='writing score',
              y='reading score',
              kind='kde',
              shade=True
)
```

```python
sns.jointplot(data =df,x='writing score',y='reading score', 
              joint_kws={'marker':'+',
                         'color':'g'},
              marginal_kws={'color':'indigo'}
             )
```

# 2. pairplot(data)

#### Plots Only among numerical data

```python
sns.pairplot(data=df)
```

```python
sns.pairplot(data=df,
             hue='gender',
)
```

```python
sns.pairplot(data=df,
             hue='gender',
             diag_kind='hist', # keeps diagonal plots as hist
             corner=True,      # Gets rid of the duplicate plots
)
```

```python
# Various Arguments

sns.pairplot(data=df,
             hue='gender',
             vars=['reading score','writing score'], # Only required values
             diag_kws={'shade':False}, # Controls the attributes of diagonal Plots
             plot_kws={'palette':'plasma'}, # Controls the attributes of off-diagonal Plots
)                               
```
