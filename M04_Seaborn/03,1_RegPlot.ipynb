```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('whitegrid')

df = sns.load_dataset('diamonds')
```

```python
diamonds = df.sample(n=200,)
```

```python
sns.regplot(x='carat',y='price',data=diamonds,)
```

```python
sns.regplot(x='carat',y='price',data=diamonds,
            fit_reg=False # or scater=False
)
```
