```python
import numpy as np

import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(".\\05-Seaborn\\StudentsPerformance.csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(10,12),dpi=200)
sns.violinplot(data=df,x='reading score',y='parental level of education	',
               hue='test preparation course',
              )

plt.legend(bbox_to_anchor=(1.2,0.5))
```

```python
plt.figure(figsize=(10,12),dpi=200)
sns.violinplot(data=df,x='reading score',y='parental level of education',
               hue='test preparation course',
               palette='spring',
               split=True,
               
              )

plt.legend(bbox_to_anchor=(1.2,0.5))
```

```python
plt.figure(figsize=(10,12),dpi=200)
sns.violinplot(data=df,x='reading score',y='parental level of education',
              inner='quartile'
              )

```

```python
plt.figure(figsize=(10,12),dpi=200)
sns.violinplot(data=df,x='reading score',y='parental level of education',
                bw=0.2
              )
```
