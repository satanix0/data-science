### Another Example:

```python
cars = sns.load_dataset('mpg').dropna()
```

```python
cars.head()
```

```python
# Only cars with 4,8 or 6 cylinders
cars = cars[cars.cylinders.isin([4, 8, 6])]
```

```python
sns.boxplot(x='mpg', data=cars)
```

```python
# U can verify with
cars.describe()['mpg']
```

```python
sns.boxplot(x='origin', y='mpg', data=cars)
```

```python
sns.boxplot(x='origin', y='mpg', data=cars,
            hue='cylinders',
            order=['japan', 'europe', 'usa'],
            width=0.4,
            linewidth=2.5,
            fliersize=8  # Control the outlier diamond size
            )
```
