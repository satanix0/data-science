```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('..\\Datasets\\dm_office_sales.csv')
df.head()
```

```python
df['division'].value_counts()
```

```python
plt.figure(figsize=(10,6),dpi=100)
sns.countplot(
    data=df,
    x='division',
)
```

```python
df['level of education'].value_counts(ascending=True).index
```

```python
plt.figure(figsize=(10,6),dpi=100)

# More Parameters
sns.countplot(
    data=df,
    x='division',
    order=['computer software','printers','office supplies','computer hardware','peripherals'],
    hue='level of education',
    hue_order=df['level of education'].value_counts(ascending=True).index,
    palette='cubehelix'
)

plt.legend(bbox_to_anchor=(1.01,0.5))
```

## Bar Plot

```python
plt.figure(figsize=(11,6),dpi=200)
sns.barplot(data=df, x = 'level of education',y='salary',
            hue='division',
            estimator=np.mean,
            ci='sd',
            order=["master's degree",'some college',
       "bachelor's degree", "associate's degree", 'high school']
           )
plt.legend(bbox_to_anchor=(1.01,0.58))
plt.tight_layout()
```
