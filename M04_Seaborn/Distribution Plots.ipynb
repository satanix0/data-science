```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('C:\\Users\\piyus\\Downloads\\Udemy Notebooks\\05-Seaborn\\dm_office_sales.csv')
```

```python
plt.figure(figsize=(12,8),dpi=100)
sns.rugplot(x='salary',data=df,height=0.228,palette='summer',hue='work experience',legend='brief')
```

```python
sns.displot(data=df,x='salary',bins=10,color='pink',edgecolor='red',linewidth=2,kde=True,)
```

```python
sns.histplot(data=df,x='salary',bins=10,color='pink',edgecolor='red',ls=':')
```

```python
sns.kdeplot(data=df,x='salary')
plt.show()
```
