```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

url ="https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/master/Python/Matplotlib/02-BarCharts/data.csv"

# Libraries needed for the tutorial
import requests
import io
    
# Downloading the csv file from your GitHub account
download = requests.get(url).content

# Reading the downloaded content and turning it into a pandas dataframe
df = pd.read_csv(io.StringIO(download.decode('utf-8')))
```

```python
ids = df['Responder_id']
langs = df['LanguagesWorkedWith']
```

```python
from collections import Counter
lang_count =Counter()
for x in langs:
    lang_count.update(x.split(';'))
    
languages = []
popularity = []
for item in lang_count.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])    
```

```python
fig,ax = plt.subplots(1,2,figsize=(10,6),dpi=100)

ax[0].barh(languages,popularity) # horizontal bars
ax[1].bar(languages,popularity)


fig.suptitle('Bar Charts',fontsize=16)
plt.tight_layout()
```

```python
py_sal = list(range(40000,60000,3000))
js_sal = list(range(35000,50000,2000))[0:7]
age = list(range(25,32))
```

```python
plt.bar(age,py_sal)
plt.bar(age,js_sal)
```

```python
# to remove the stacking up of graphs
width = 0.4
x_indexs = np.arange(len(age))
plt.bar(x_indexs+width,py_sal,width=width,label='Python Devs')
plt.bar(x_indexs,js_sal,width=width,label='JS Devs')
plt.legend()

plt.xticks(ticks=x_indexs,labels=age)
plt.tight_layout()
plt.xlabel('age')
plt.ylabel('Salary')
plt.title('Salary Comparison')
```
