```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
x1= np.random.randn(1500)
x2= np.random.randn(1500)
```

```python
np.column_stack((x1, x2))
```

```python
px.scatter(x=x1, y=x2, title="Scatter Plot").show()
```

```python
from sklearn.ensemble import IsolationForest
clf = IsolationForest(n_estimators=100, max_samples=1000, contamination=0.09, n_jobs=-1, random_state=42)
y = clf.fit_predict(np.column_stack((x1, x2)))
```

```python
pd.DataFrame(y).value_counts()
```

```python
px.scatter(x=x1, y=x2, color=y, title="Outliers Detected using Isolation Forest").show()
```
