## Silhouette Score for Clustering

The **Silhouette Score** is a metric used to evaluate the quality of clustering by measuring how similar an object is to its own cluster (cohesion) compared to other clusters (separation).<br>
It ranges from -1 to +1, where a high value indicates that the object is well-matched to its own cluster and poorly matched to neighboring clusters.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_blobs
X, _ = make_blobs(n_samples=1500, centers=4, cluster_std=1.5, random_state=42,)
X.shape
```

```python
sns.scatterplot(x=X[:, 0], y=X[:, 1])
```

```python
def plot_labels(data, model):
    labels = model.labels_
    sns.scatterplot(x=data[:, 0], y=data[:, 1], hue=labels, palette='Set1')
    
```

```python
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.metrics import silhouette_score

# Fit KMeans with a specified number of clusters
kmeans = KMeans(n_clusters=4, random_state=42)
cluster_labels = kmeans.fit_predict(X)

plot_labels(X, kmeans)

# Calculate Silhouette Score
score = silhouette_score(X, cluster_labels)
print("Silhouette Score for KMeans:", score)
```

```python
agg = AgglomerativeClustering(n_clusters=4)
cluster_labels = agg.fit_predict(X)

plot_labels(X, agg)

# Calculate Silhouette Score
score = silhouette_score(X, cluster_labels)
print("Silhouette Score for Agglomerative:", score)
```

```python
dbscan = DBSCAN(eps=1, min_samples=5)
cluster_labels = dbscan.fit_predict(X)

plot_labels(X, dbscan)

# Calculate Silhouette Score
score = silhouette_score(X, cluster_labels)
print("Silhouette Score for DBSCAN:", score)
```
