### Q1. Find all the unique 'Wind Speed' values in the data. 

```python
df['Wind Speed_km/h'].unique()
```

### Q2. Find the number of times when the 'Weather is exactly Clear'.

```python
# using value_counts()
df['Weather'].value_counts()['Clear']
```

```python
# using groupby()
df.groupby(by='Weather').count().loc['Clear']
```

```python
# using conditional filtering()
df[df['Weather']=='Clear'].count()
```

### Q3. Find the number of times when the 'Wind Speed was exactly 4 km/h'.

```python
df['Wind Speed_km/h'].value_counts()[4]
```

```python
df[df['Wind Speed_km/h']==4]
```

### Q11. What is the Mean value of each column against each 'Weather Conditon'
