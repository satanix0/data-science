```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
df_genre = pd.read_csv('SpotifyFeatures.csv')
df_genre.info()
```

```python
df_genre.head()
```

```python
df_genre['duration']=np.round(df_genre['duration_ms']/(1000*60),2)
```

```python
plt.figure(figsize=(12,6))
plt.title('Mean duration of the songs of diferent genere',fontdict={'fontsize':20})
sns.barplot(x='duration',y='genre',data=df_genre,
            palette='rainbow',estimator=np.mean
           )
plt.xlabel(xlabel='Duration (in minutes)')
plt.show()
```

## Top 5 generes w.r.t popularity

```python
popular_gen=df_genre.groupby(by='genre').sum()['popularity'].sort_values(ascending=False)
```

```python
popular_gen = popular_gen.reset_index()
```

```python
plt.figure(figsize=(12,6))
plt.title('Popularity of different generes',fontdict={'fontsize':20})
sns.barplot(x='popularity',y='genre',data=popular_gen[0:5],
            palette='rainbow',
           )
plt.xlabel(xlabel='Genre')
plt.show()
```
