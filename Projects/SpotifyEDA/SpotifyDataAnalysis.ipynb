```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading Artist
df_tracks = pd.read_csv('tracks.csv')
df_tracks.info()
```

```python
df_tracks.isnull().sum() # Checking for Null values
```

### A1: 10 Least Popular Songs.

```python
sorted_df = df_tracks.sort_values(by='popularity',ascending=True)
sorted_df.head(10)
```

```python
# Statistical Description

df_tracks.describe().transpose()
```

### A2. Top 10 Songs with Popularity Greater than 90.

```python
df_tracks[df_tracks['popularity']> 90][0:11].sort_values(by='popularity',ascending=False)
```

```python
## Changing Index on Release date

df_tracks.set_index('release_date',inplace=True)
df_tracks.index=pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
# Changing duration, from ms to mintutes

df_tracks['duration_min'] = np.round(df_tracks['duration_ms'] /(1000*60),2)

df_tracks.drop(labels=['duration','duration_s','duration_ms'],axis=1,inplace=True)
```

### Creating a correlation map using Seaborn among all columns except columns ['key','mode','explicit']

```python
corr_df = df_tracks.drop(labels=['key','mode','explicit'],axis=1).corr()
fig=plt.figure(figsize=(10,5),dpi=100)
sns.heatmap(data=corr_df,
            annot=True,
            cmap='plasma',
            linewidth=2,
            linecolor='black',
            fmt='.1g',
            vmin=-1,
            vmax=1,
           )
plt.title(label='Correlation map',
          fontdict={'fontsize':20,'fontweight':'bold'})
```

### Creating a regression plot, using Seaborn, between columns 'loudness' and 'energy' as they have a positive correlation

```python
fig=plt.figure(figsize=(10,5),dpi=100)
# taking a small random sample pf 5% of total data

sample_data = df_tracks.sample(frac=0.005)

sns.regplot(data=sample_data,y='loudness',x='energy',
            scatter_kws={'alpha':0.2,
                         'color':'salmon'
                        },
            line_kws={'linewidth':3,
                    'color':'green'
                     }
           )
plt.title(label='Loudness vs Energy Correlation',
          fontdict={'fontsize':20,'fontweight':'bold'})
```

### Creating a regression plot, using Seaborn, between columns 'popularity' and 'acousticness' as they have a positive correlation

```python
fig=plt.figure(figsize=(10,5),dpi=100)
# taking a small random sample pf 5% of total data

sample_data = df_tracks.sample(frac=0.005)

sns.regplot(data=sample_data,y='popularity',x='acousticness',
            scatter_kws={'alpha':1,
                         'color':'orange',
                        },
            line_kws={'linewidth':3,
                    'color':'purple'
                     }
           )
plt.title(label='Popularity Correlation',
          fontdict={'fontsize':20,'fontweight':'bold'})
```

### Visualising the Number of songs released each year.

```python
# Extracting year from the release date.
df_tracks['release_year']= df_tracks.index.year
```

```python
plt.figure(figsize=(25,8))
sns.countplot(data=df_tracks,x='release_year')
plt.xticks(rotation=90)
plt.show()
plt.tight_layout()
```
