```python
import numpy as np

import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("..//05-Seaborn//StudentsPerformance.csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(9,4))
sns.boxplot(data=df,y='reading score',x='parental level of education',
            hue='test preparation course',
            palette='inferno'
           
)
plt.tight_layout()
```

### Another Example:

```python
cars = sns.load_dataset('mpg').dropna()
```

```python
cars.head()
```

```python
# Only cars with 4,8 or 6 cylinders
cars=cars[cars.cylinders.isin([4,8,6])] 
```

```python
sns.boxplot(x='mpg',data=cars)
```

```python
# U can verify with 
cars.describe()['mpg']
```

```python
sns.boxplot(x='origin',y='mpg',data=cars)
```

```python
sns.boxplot(x='origin',y='mpg',data=cars,
            hue='cylinders',
            order=['japan','europe','usa'],
            width=0.4,
            linewidth=2.5,
            fliersize=8 # Control the outlier diamond size
)
```
