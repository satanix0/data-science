```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')

df = pd.read_csv('C://Users//piyus//Downloads//Udemy Notebooks//05-Seaborn//StudentsPerformance.csv')
```

```python
df.head()
```

# catplot()

```python
sns.catplot(data=df,x='gender',y='math score',
            kind='box',
            row='lunch'
           )
plt.show()
```

```python
sns.catplot(data=df,x='gender',y='math score',
            kind='box',
            col='lunch',
            row='test preparation course'
           )
plt.show()
```

### PairGrid - Used to map different plots in a pairplot

```python
g = sns.PairGrid(df)

# g=g.map_upper(sns.scatterplot,color='g')
# g=g.map_diag(sns.histplot,color='black')
# g=g.map_lower(sns.kdeplot,color='indigo')
# plt.show()
```
