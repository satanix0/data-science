```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='darkgrid')
```

```python
diamonds= sns.load_dataset('diamonds')
```

```python
diamonds = diamonds.sample(n=200) # taking sample of 200
diamonds.head()
```

```python
sns.regplot(data=diamonds,x='carat',y='price',)
```

```python
sns.regplot(data=diamonds,x='carat',y='price',
            fit_reg=False,          
           )
```

```python
plt.xlim((0,2.6))
sns.regplot(data=diamonds,x='carat',y='price',
            scatter=False
           )
```

```python
cut_map={
    "Fair":1,
    "Good":2,
    "Very Good":3,
    "Premium":4,
    "Ideal":4,
}
diamonds['cut_value']=diamonds['cut'].map(arg=cut_map)

diamonds['cut_value'].value_counts()
```

```python
sns.regplot(data=diamonds,x='cut_value',y='price',
           
           )
```

```python
sns.regplot(data=diamonds,x='cut_value',y='price',
            x_jitter=0.1
           )
```

```python
sns.regplot(data=diamonds,x='cut_value',y='price',
            x_estimator=np.mean,
           )
```
